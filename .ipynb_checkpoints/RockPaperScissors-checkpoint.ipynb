# Ween's Rock Paper Scissors Game


    


### prompts the user to pick between single player where a bot randomly throws rock paper or scissors while the user gets to choose
### two player, where the user can play against another user, their throw is hidden from the other user
### multiplayer where an unlimited number of players can play. can only win if they throw 2 total of the rock paper or scissors. 
###    cannot throw all three or else it is a tie. there is a winning team and losing team."""

```python
import getpass
import random
import ListString

class RockPaperScissorsGame():
    
    def __init__ (self):
        
        """" the constructors of each gamemode play one round of the game
             each game mode is explained in the notebook """
        
        game = True
        while game:
            
            throw_one = getpass.getpass(prompt = "Player 1: Rock, Paper, or Scissors? ")
            throw_two = getpass.getpass(prompt = "Player 2: Rock, Paper, or Scissors? ")
            outcome = self.evaluate(throw_one, throw_two)
           
            if outcome == "Player 1":
                print("Player 1 threw " + throw_one + " and Player 2 threw " + throw_two + ": Player 1 wins!")
                game = False
            elif outcome == "Player 2":
                print("Player 1 threw " + throw_one + " and Player 2 threw " + throw_two + ": Player 2 wins!")
                game = False
            elif outcome == "Tie":
                print("It's a tie! Both Players threw " + throw_one + " Throw again!")
            else:
                print("Oops something went wrong, try again!")
                
    def evaluate(self, throw_one, throw_two):
        
        """" code for evaluating which throw will win of two throws
             integral to each game mode and thus is in the superclass which other gamemodes inherit from """
        
        winner = ""
       
        if throw_one.lower() == throw_two.lower():
            winner = "Tie"
        elif throw_one.lower() == "rock":
            if throw_two.lower() == "paper":
                winner = "Player 2"
            elif throw_two.lower() == "scissors":
                winner = "Player 1"
        elif throw_one.lower() == "paper":
            if throw_two.lower() == "scissors":
                winner = "Player 2"
            elif throw_two.lower() == "rock":
                winner = "Player 1"
        elif throw_one.lower() == "scissors":
            if throw_two.lower() == "rock":
                winner = "Player 2"
            elif throw_two.lower() == "paper":
                winner = "Player 1"
        return winner
    
class SinglePlayerGame(RockPaperScissorsGame):
    
    def __init__(self):
    
        """" constructor for the single player game mode, with a bot to pick for player 2 """
    
        game = True
        bot_choices = ["Rock", "Paper", "Scissors"]
        
        while game:
            
            throw_one = input("Rock, Paper, or Scissors? ")
            throw_two = random.choice(bot_choices)
            outcome = self.evaluate(throw_one, throw_two)
            
            if outcome == "Player 1":
                print("You threw " + throw_one + " and the computer threw " + throw_two + ": You win!")
                game = False
            elif outcome == "Player 2":
                print("You threw " + throw_one + " and the computer threw " + throw_two + ": The computer wins!")
                game = False
            elif outcome == "Tie":
                print("It's a tie! You and the computer both threw " + throw_one + ". Throw again!")
            else:
                print("Oops something went wrong, try again!")
                
class MultiPlayerGame(RockPaperScissorsGame):
    
    def __init__ (self):
        
        """"constructor which runs the multiplayer game mode"""
        
        throws = {}
        player_count = int(input("How many players are there? :\t"))
        game = True
        
        while game:


            first = ""
            second = ""
            winners = []
            losers = []
            winning_throw = ""
            ##losing_throw = ""
            tie = False
            tracer = 1

            while tracer <= player_count:
                player_throw = getpass.getpass(prompt = "Player " + str(tracer) + ": Rock, Paper, or Scissors? ")
                throws[tracer] = player_throw
                tracer += 1
            
            first = throws[1]
            second = ""
            
            for item in throws:
                if throws[item] == first:
                    game = True
                elif second == "" or throws[item] == second:
                    second = throws[item]
                else:
                    print("It's a draw! Rock, Paper, and Scissors were each thrown at least once. Try again.")
                    tie = True
            if second == "":
                print("It's a draw! Everybody threw " + first + ". Try again!")
                tie = True
            
            if tie:
                continue
            outcome = self.evaluate(first, second)
            if outcome == "Player 1":
                winning_throw = first.lower()
            else:
                winning_throw = second.lower()
            for thing in throws:
                if throws[thing] == winning_throw.lower():
                    winners.append(thing)
                else:
                    losers.append(thing)

            print("The winners are Player " + ListString.list_to_string(self, winners, ", Player ") + " and the losers are Player " + ListString.list_to_string(self, losers, ", Player ") + ". Good Game!")
            game = False
            
```
